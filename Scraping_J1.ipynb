```python
# recuperer tout les films 
url = "https://www.imdb.com/search/title/?groups=top_250&sort=user_rating"
r = requests.get(url)

soup = BeautifulSoup(r.text,"html.parser")

data = [] 

for film in soup.find_all('div', class_="lister-item-content"):
    rank = int(film.find('span', class_='lister-item-index').text.replace('.',''))
    title = film.find('a').text
    time = int(film.find('span',class_='runtime').text.replace(" min", ''))
    genre = str(film.find('span', class_='genre').text.strip())
    score = eval(film.find('div', class_='ratings-imdb-rating')['data-value'])
    year = eval(film.find('span', class_='lister-item-year').text)

    dico = {'rank':rank,
            'title': title,
            'time': time,
            'genre': genre,
            'score': score,
            'year': year,}
    
    data.append(dico)
    
```
